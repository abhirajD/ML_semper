```python
# Import the library
from sklearn import tree
```

```python
# We will work with a classic example dataset - Iris
from sklearn.datasets import load_iris
```

```python
# We load the dataset ...
iris = load_iris()
```

```python
print(iris)
```

```python
# Here we build the Decision Tree classifier
clf = tree.DecisionTreeClassifier()
clf = clf.fit(iris.data, iris.target)
```

```python
print(iris.data)
```

```python
print(iris.target)
```

```python
import graphviz
```

```python
dot_data = tree.export_graphviz(clf, out_file=None)
```

```python
graph = graphviz.Source(dot_data)
```

```python
dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=iris.feature_names,  
                         class_names=iris.target_names,  
                         filled=True, rounded=False,  
                         special_characters=True)
```

```python
graph = graphviz.Source(dot_data)
```

```python
graph
```

```python
# Here we predict a new value
clf.predict(iris.data[:1000, :])
```

```python
# This will give us the probability for each class - it is the fraction of training samples in each leaf
clf.predict_proba(iris.data[:1000, :])
```
